```python
import numpy as np
import cv2
```

```python
def clip(video, starts, ends):
    """Get clip between start and end
    
    Parameters
    ----------
    video: VideoCapture
        Capture file
    start: float64
        Timestamp in milisecond
    end: float64
        Timestamp in milisecond
    """
    
    clips = [] # Array to keep clips
    clip_number = 0
    n_clips = len(starts)
    
    # frames = []
    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*"XVID")
    out = cv2.VideoWriter("clips/clip-0.avi", fourcc, 30.0, (490, 360))

    while clip_number < n_clips:
        ret, frame = video.read()
        
        cur_timestamp = video.get(cv2.CAP_PROP_POS_MSEC)

        if cur_timestamp >= starts[clip_number] and cur_timestamp <= ends[clip_number]:
            if clip_number > 0:
                print(np.max(frame))
                break
            out.write(frame)
            
        if cur_timestamp > ends[clip_number]:
            clip_number += 1
            # clips.append(np.array(frames))
            # frames = []
            fourcc = cv2.VideoWriter_fourcc(*"XVID")
            out = cv2.VideoWriter("clips/clip-{}.avi".format(clip_number),fourcc, 30.0, (360, 490))
            
            
    # assert len(clips) == n_clips, "Number of clips must be equal to number of input clips."
    out.release()
    video.release() 
    # return clips
    
```
